```python
from math import pi, cos, sin, tan, atan2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.image as mpimg
%matplotlib inline

plt.style.use('seaborn')
plt.rcParams['legend.numpoints'] = 1
plt.rcParams['axes.labelsize'] = 20
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.rcParams['legend.fontsize'] = 10

legend_opts = {
    'fontsize':15,
    'frameon':True, 
    'framealpha':1,
    'facecolor':'white',
    'edgecolor':'black',
    'labelspacing':0.1}

def savefig(fig, filename, **kwargs):
    fig.savefig(f'../viz/{filename}', bbox_inches='tight', **kwargs)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.arrow(0, 0, 1, 2, length_includes_head=True, head_width=0.2, fc='tab:red', ec='tab:red')
ax.arrow(0, 0, -2, 3, length_includes_head=True, head_width=0.2, fc='tab:red', ec='tab:red')
ax.arrow(0, 0, -1, 4, length_includes_head=True, head_width=0.2, fc='tab:green', ec='tab:green')
ax.plot([1, 5], [2, 10], '--', lw=1, color='tab:red')
ax.plot([-2, -5], [3, -5*-3/2], '--', lw=1, color='tab:red')
ax.fill_between(x=[-5, 0, 5], y1=[-5*-3/2, 0, 10], y2=[10, 10, 10], color='tab:red', alpha=0.2)

ax.axhline(0, color='tab:gray', lw=1)
ax.axvline(0, color='tab:gray', lw=1)

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xticks(np.arange(-5, 5.1, 1))
ax.set_yticks(np.arange(-5, 5.1, 1))
# savefig(fig, 'ab.png')
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.arrow(0, 0, 2, 3, length_includes_head=True, head_width=0.2, fc='tab:pink', ec='tab:pink')
ax.arrow(0, 0, -1, -4, length_includes_head=True, head_width=0.2, fc='tab:olive', ec='tab:olive')
ax.axhline(0, color='tab:gray', lw=1)
ax.axvline(0, color='tab:gray', lw=1)

angle = 0.5
y1 = 3*cos(angle)
y2 = 3*sin(angle)
y_arrow = plt.Arrow(0, 0, y1, y2, fc='black', ec='black', width=0.5)
ax.add_patch(y_arrow)
ax.plot([10*cos(angle+pi/2), 10*cos(angle-pi/2)], [10*sin(angle+pi/2), 10*sin(angle-pi/2)], ls='--', lw='1', color='tab:cyan')
ax.fill_between([-10, 10], [-10*tan(angle+pi/2), 10*tan(angle-pi/2)], [10, 10], color='tab:pink', alpha=0.2)
ax.fill_between([-10, 10], [-10*tan(angle+pi/2), 10*tan(angle-pi/2)], [-10, -10], color='tab:olive', alpha=0.2)

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xticks(np.arange(-5, 5.1, 1))
ax.set_yticks(np.arange(-5, 5.1, 1))
# savefig(fig, 'y.png')
```

```python
def update(i, fig, ax, verbose=True):
    if verbose and i%10==0:
        print(i)
    ax.collections = [ax.collections[0]]
    ax.lines = ax.lines[:4]
    ax.patches = []
    if len(ax.texts) > 3:
        ax.texts = ax.texts[:3]
    angle = angles[i]
    y1 = 3*cos(angle)
    y2 = 3*sin(angle)
    y_arrow = plt.Arrow(0, 0, y1, y2, fc='black', ec='black', width=0.5)
    ax.add_patch(y_arrow)
    ax.plot([10*cos(angle+pi/2), 10*cos(angle-pi/2)], [10*sin(angle+pi/2), 10*sin(angle-pi/2)], ls='--', lw='1', color='tab:cyan')
    ax.fill_between([-10, 10], [-10*tan(angle+pi/2), 10*tan(angle-pi/2)], [10, 10], color='tab:pink', alpha=0.2)
    ax.fill_between([-10, 10], [-10*tan(angle+pi/2), 10*tan(angle-pi/2)], [-10, -10], color='tab:olive', alpha=0.2)
    ax.text(y1, y2, r'$\vec{y}$', fontsize=15, ha='right', va='bottom')
```

```python
angles = np.linspace(atan2(3, -2)-pi/2, atan2(2, 1)+pi/2, 50)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.arrow(0, 0, 1, 2, length_includes_head=True, head_width=0.2, fc='tab:red', ec='tab:red')
ax.text(1.3, 2, r'$\vec{a_1}$', fontsize=15, ha='left', va='bottom', color='tab:red')
ax.arrow(0, 0, -2, 3, length_includes_head=True, head_width=0.2, fc='tab:red', ec='tab:red')
ax.text(-2.3, 2.7, r'$\vec{a_2}$', fontsize=15, ha='right', va='bottom', color='tab:red')
ax.arrow(0, 0, -1, 4, length_includes_head=True, head_width=0.2, fc='tab:green', ec='tab:green')
ax.text(-1, 4, r'$\vec{b}$', fontsize=15, ha='right', va='bottom', color='tab:green')
ax.plot([1, 5], [2, 10], '--', lw=1, color='tab:red')
ax.plot([-2, -5], [3, -5*-3/2], '--', lw=1, color='tab:red')
ax.axhline(0, color='tab:gray', lw=1)
ax.axvline(0, color='tab:gray', lw=1)
ax.fill_between(x=[-5, 0, 5], y1=[-5*-3/2, 0, 10], y2=[10, 10, 10], color='tab:red', alpha=0.2)


ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xticks(np.arange(-5, 5.1, 1))
ax.set_yticks(np.arange(-5, 5.1, 1))
plt.show()

anim = FuncAnimation(fig, lambda i: update(i, fig, ax, verbose=False), 
                     frames=range(len(angles)), interval=50, repeat=True)
anim.save('../viz/y_sweep.mp4', writer='ffmpeg')
```
